```python
# Importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the dataframe

# df = pd.read_csv("E:Gaurav CS/Data for Analysis/diabetes.csv")
df = pd.read_csv("diabetes.csv")
```

```python
# dividing the data into the independent and dependent variable

x = df.drop('Outcome', axis=1)
y = df['Outcome']
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(3,3))
sns.distplot(x['Age'],bins=70)
plt.title('Age wise distribution of Diabetic patients')
plt.xlabel('Age')
plt.ylabel('density of population age wise')
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.countplot(x,x='Pregnancies')
plt.title('Pregnancy count of Diabetic patients')
plt.xlabel('Number of Pregnancies')
plt.ylabel('count')
plt.show()
```

```python
# splitting the dataset

from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test = tts(x,y,train_size=0.2,random_state=0)
```

```python
x_train.head()
```

```python
df['Outcome'].value_counts()
```

```python
# Training the dataset
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_test_pred = lr.predict(x_test)
```

```python
y_test_pred
```

```python
y_train_pred = lr.predict(x_train)
```

```python
y_train_pred
```

```python
cm_test = metrics.confusion_matrix(y_test, y_test_pred)
cm_test
```

```python
acs_test = metrics.accuracy_score(y_test, y_test_pred)
acs_test
```

```python
acs_train = metrics.accuracy_score(y_train, y_train_pred)
acs_train
```

```python
y_test = np.array(y_test)
y_test
```
